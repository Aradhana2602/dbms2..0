```python
df=pd.read_csv("student.csv")
```

```python
#PART A : STUDENT DATASET
```

```python
df.describe()
```

```python
grouped_stats = df.groupby("gender").agg(
    math_mean=("math score","mean"),
    math_median=("math score","median"),
    math_min=("math score","min"),
    math_max=("math score","max"),
    math_std=("math score","std"),

    reading_mean=("reading score","mean"),
    readin_std=("reading score","std"),

    writing_mean=("writing score","mean"),
    writing_std=("writing score","std")
)

print(grouped_stats)
```

```python
math_score_list = df.groupby("gender")["math score"].apply(list)
print(math_score_list)
```

```python
for g in df["gender"].unique():
    print ("\n Statistical details for :",g)
    print (df[df["gender"] == g][["math score","reading score","writing score"]].describe())
```

```python
# Load iris dataset
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Group by species and get descriptive statistics
grouped_stats = df.groupby('Species').describe()
print(grouped_stats)
```
